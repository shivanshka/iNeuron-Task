# Resume Parser

```python
import re
import glob
import PyPDF2 as pdf
import pandas as pd
from docx import Document
```

```python
path = r"E:\Shivansh\iNeuron\Jupyter\Data\resume\same-resume-year-wise-master\*.*"
file_list = glob.glob(path)
```

```python
name=[]
email=[]
linked =[]
git=[]
skill=[]
```

```python
def extracter(file,txt):
    match=re.split("[\\\]",file)[-1]
    name = match.replace(re.findall("\.[a-z]{2,4}$",match)[0],"")
    
    email = re.findall('[a-zA-Z0-9_\-\.]+@[a-z]+[\.][a-z]{2,3}',txt)
    if email == []:
        email = "NaN"
    else:
        email = email[0]
    
    linked = re.findall('linkedin.com/[a-zA-Z0-9\\-\.]+\s',txt)
    if linked == []:
        linked = "NaN"
    else:
        linked = linked[0]
        
    git = re.findall('github.com/[a-zA-Z0-9\\-\.]+\s',txt)
    if git == []:
        git="NaN"
    else:
        git = git[0]
    return name,email,linked,git
```

```python
def doc_text(i):
    doc = Document(i)
    s=""
    for i in doc.paragraphs:
        if(len(i.text)>1):
            s +=i.text
    return s
```

```python
for i in file_list:
    if i.endswith(".pdf"):
        file = pdf.PdfFileReader(i)
        data = file.getPage(0).extractText()
        
    elif i.endswith(".docx"):
        data = doc_text(i)
        
    n,e,l,g = extracter(i,data)
    name.append(n)
    email.append(e)
    linked.append(l)
    git.append(g)
```

```python
data2 = {"email_id" : email, "linked_id": linked, "github_id": git}
```

```python
df = pd.DataFrame(data2,index=name)
```
